# 課題 1

以下の 5 つの関数について、$1\leqq x \leqq 8$の区間のグラフを重ねて描画する

- $y = x$
- $y = x^2$
- $y = 2^x$
- $y = logx$
- $y = xlogx$


```python
import matplotlib.pyplot as plt
import math
```

```python
xs: list[float] = []
ys_x: list[float] = []
ys_x2: list[float] = []
ys_2x: list[float] = []
ys_logx: list[float] = []
ys_xlogx: list[float] = []

for i in range(10, 81):
    x = i / 10
    xs.append(x)
    ys_x.append(x)
    ys_x2.append(x ** 2)
    ys_2x.append(2 ** x)
    ys_logx.append(math.log(x))
    ys_xlogx.append(x * math.log(x))

plt.plot(xs, ys_x, label='$x$')
plt.plot(xs, ys_x2, label='$x^2$')
plt.plot(xs, ys_2x, label='$2^x$')
plt.plot(xs, ys_logx, label='$log(x)$')
plt.plot(xs, ys_xlogx, label='$xlog(x)$')

plt.title('Graphs of several functions')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(1, 10)

plt.legend()
plt.show()
```

## それぞれの関数の増加の特性の比較

- $y=x$
  x が増えると y も等比例に増加する
- $y=x^2$
  x が増えると y はより急速に増加し、曲線的に上昇する
- $y=2^x$
  x が増えると y は指数的に急速に増加する
- $y=logx$
  x が増えても y の増加は緩やかで。x が大きくなっても y は比較的遅いペースで増加する
- $y=xlogx$
  x が増えると y も増加するが、x が大きくなるほど y の増加は鈍化する
