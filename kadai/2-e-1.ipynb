# 課題1
以下の5つの関数について、$1\leqq x \leqq 8$の区間のグラフを重ねて描画する
- $y = x$
- $y = x^2$
- $y = 2^x$
- $y = logx$
- $y = xlogx$

```python
import matplotlib.pyplot as plt
import math
```

```python
xs = []
ys_x = []
ys_x2 = []
ys_2x = []
ys_logx = []
ys_xlogx = []

for i in range(10, 81):
    x = i / 10
    xs.append(x)
    ys_x.append(x)
    ys_x2.append(x ** 2)
    ys_2x.append(2 ** x)
    ys_logx.append(math.log(x))
    ys_xlogx.append(x * math.log(x))

plt.plot(xs, ys_x, label='$x$')
plt.plot(xs, ys_x2, label='$x^2$')
plt.plot(xs, ys_2x, label='$2^x$')
plt.plot(xs, ys_logx, label='$log(x)$')
plt.plot(xs, ys_xlogx, label='$xlog(x)$')

plt.title('Graphs of several functions')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.show()
```

## それぞれの関数の増加の特性の比較
- $y=x$
xが増えるとyも等比例に増加する
- $y=x^2$
xが増えるとyはより急速に増加し、曲線的に上昇する
- $y=2^x$
xが増えるとyは指数的に急速に増加する
- $y=logx$
xが増えてもyの増加は緩やかで。xが大きくなってもyは比較的遅いペースで増加する
- $y=xlogx$
xが増えるとyも増加するが、xが大きくなるほどyの増加は鈍化する
