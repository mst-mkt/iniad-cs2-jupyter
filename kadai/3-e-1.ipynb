```python
import matplotlib.pyplot as plt
import timeit
```

```python
def f3(n: int) -> int:
    i = 1
    s = 0
    while i <= n:
        s += 1
        i *= 2
    return s
```

```python
num_iter = 1000
x = [1.3 ** i for i in range(0, 100)]
y = []
for n in x:
    time = timeit.timeit('f3(n)', globals=globals(), number=num_iter)
    y.append(time / num_iter)
```

```python
plt.plot(x, y, 'o')

plt.title('Time of f3(n)')
plt.xlabel('n')
plt.ylabel('Average time (s)')

plt.show()
```
