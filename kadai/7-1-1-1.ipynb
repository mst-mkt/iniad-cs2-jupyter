```python
import matplotlib.pyplot as plt
```

```python
def simple_interest(principal: int, rate: float, years: int) -> int:
    return int(principal * (1 + rate * years))

def compound_interest(principal: int, rate: float, years: int) -> int:
    return int(principal * (1 + rate) ** years)
```

```python
principal = 1_000_000
rate = 0.02
years = 50

xs: list[int] = list(range(years + 1))
ys_simple: list[int] = [simple_interest(principal, rate, x) for x in xs]
ys_compound: list[int] = [compound_interest(principal, rate, x) for x in xs]
```

```python
plt.plot(xs, ys_simple, label="Simple")
plt.plot(xs, ys_compound, label="Compound")

plt.title("Simple and Compound interest")
plt.xlabel("Year")
plt.ylabel("Amount")

plt.legend()
plt.show()
```
