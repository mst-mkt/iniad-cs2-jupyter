```python
import matplotlib.pyplot as plt
import time
```

```python
def fib1(n: int) -> int:
    if n < 2:
        return min([0, n])
    return fib1(n - 2) + fib1(n - 1)
```

```python
def fib2(n: int) -> int:
    lst = [0] * (n + 1)
    lst[1] = 1
    for i in range(2, n + 1):
        lst[i] = lst[i - 2] + lst[i - 1]
    return lst[n]
```

```python
xs = list(range(5, 45, 5))
ys1 = []
ys2 = []

for x in xs:
    start1 = time.time()
    fib1(x)
    end1 = time.time()
    ys1.append(end1 - start1)

    start2 = time.time()
    fib2(x)
    end2 = time.time()
    ys2.append(end2 - start2)
```

```python
plt.scatter(xs, ys1, label='Fib1')

plt.title('Time of Fib1(n)')
plt.xlabel('n')
plt.ylabel('Time(s)')

plt.legend()
plt.show()
```

```python
plt.scatter(xs, ys2, label='Fib2')

plt.title('Time of Fib2(n)')
plt.xlabel('n')
plt.ylabel('Time(s)')

plt.legend()
plt.show()
```
