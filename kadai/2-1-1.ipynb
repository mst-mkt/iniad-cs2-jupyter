```python
import matplotlib.pyplot as plt
import time
import random as r
```

```python
def linear_search(lst: list[int], x: int) -> bool:
    for item in lst:
        if item == x:
            return True
    return False
```

```python
def binary_search(lst: list[int], x: int) -> bool:
    start, end = 0, len(lst)
    while start < end:
        center = start + (end - start) // 2
        if lst[center] == x:
            return True
        elif lst[center] > x:
            end = center
        else:
            start = center + 1
    return False
```

```python
xs = [1, 10, 100, 1000, 10000]
ys_linear = []
ys_binary = []

for x in xs:
    target = r.randint(0, x)

    start_linear = time.time()
    linear_search(list(range(x)), target)
    end_linear = time.time() - start_linear

    start_binary = time.time()
    binary_search(list(range(x)), target)
    end_binary = time.time() - start_binary

    ys_linear.append(end_linear * 1000)
    ys_binary.append(end_binary * 1000)

plt.scatter(xs, ys_linear, label="linear search")
plt.scatter(xs, ys_binary, label="binary search", marker="x")

plt.title("Linear Search and Binary Search")
plt.xlabel("Length")
plt.ylabel("Time (ms)")

plt.legend()
plt.show()
```
