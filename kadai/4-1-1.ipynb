```python
import matplotlib.pyplot as plt
import time
```

```python
def find_center(lst: list[int]) -> int:
    center = len(lst) // 2
    return lst.index(center)
```

```python
xs = list(map(lambda x: x * 100_000, range(1, 11)))
ys: list[float] = []

for x in xs:
    start = time.time()
    find_center(list(range(x)))
    end = time.time()
    ys.append(end - start)

plt.plot(xs, ys, 'o')

plt.title('Time of find_center(lst)')
plt.xlabel('Length')
plt.ylabel('Time (s)')

plt.show()
```
