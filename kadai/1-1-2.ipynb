```python
import matplotlib.pyplot as plt
```

```python
xs: list[int] = []
ys_simple: list[float] = []
ys_compound: list[float] = []

principal = 10000
rate = 0.03

for x in range(21):
    xs.append(x)
    ys_simple.append((x * principal * rate) + principal)
    ys_compound.append(principal * (1 + rate) ** x)

plt.plot(xs, ys_simple, label="Simple")
plt.plot(xs, ys_compound, label="Compound")

plt.title("Simple and Compound interest")
plt.xlabel("Years")
plt.ylabel("Value")

plt.legend()
plt.show()
```
