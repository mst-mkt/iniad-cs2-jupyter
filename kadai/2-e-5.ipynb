```python
import matplotlib.pyplot as plt
import csv
```

```python
months = []
rainfalls = []

with open('sample/weather_tokyo_2020.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        months.append(row[0])
        rainfalls.append(float(row[4]))

plt.bar(months, rainfalls)

plt.title('Tokyo monthly rainfall in 2020')
plt.xlabel('Month')
plt.ylabel('mm')

plt.legend(['Rainfall'])
plt.show()
```
