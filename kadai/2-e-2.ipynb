```python
import matplotlib.pyplot as plt
```

```python
xs: list[float] = []
ys: list[float] = []

for i in range(-100, 101):
    x = i / 10
    xs.append(x)
    if (i != 0):
        ys.append(1 / x)
    else:
        ys.append(float('nan'))

plt.plot(xs, ys)

plt.title('Graph of y = 1/x')
plt.xlabel('x')
plt.ylabel('y')

plt.legend(['y = 1/x'])
plt.show()
```
