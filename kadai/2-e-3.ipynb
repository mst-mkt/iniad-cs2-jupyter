```python
import matplotlib.pyplot as plt
import csv
```

```python
xs: list[str] = []
ys: list[float] = []

with open('sample/weather_tokyo_2020.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        xs.append(row[0])
        ys.append(float(row[3]))

plt.plot(xs, ys)

plt.title('Tokyo monthly mean temperature in 2020')
plt.xlabel('Month')
plt.ylabel('â„ƒ')

plt.legend(['Temp.'])
plt.show()
```
