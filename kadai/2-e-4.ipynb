```python
import matplotlib.pyplot as plt
import csv
```

```python
month = []
temp_max = []
temp_min = []
temp_avg = []

with open('sample/weather_tokyo_2020.csv', 'r') as f:
    render = csv.reader(f)
    for row in render:
        month.append(row[0])
        temp_max.append(float(row[1]))
        temp_min.append(float(row[2]))
        temp_avg.append(float(row[3]))

plt.plot(month, temp_max, label='Highest Temp.')
plt.plot(month, temp_min, label='Lowest Temp.')
plt.plot(month, temp_avg, label='Ave Temp.', linestyle='dashed')

plt.title('Tokyo Monthly Temperature in 2020')
plt.xlabel('Month')
plt.ylabel('')

plt.legend()
plt.show()
```
