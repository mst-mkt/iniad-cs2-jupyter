```python
import matplotlib.pyplot as plt
import time
```

```python
def selection_sort(lst: list[int]) -> None:
    length = len(lst)
    for i in range(length):
        min_index = i
        for j in range(i + 1, length):
            if lst[j] < lst[min_index]:
                min_index = j
        lst[i], lst[min_index] = lst[min_index], lst[i]
```

```python
xs = list(range(100, 1001, 100))
ys = []

for x in xs:
    lst = list(range(x))
    start_time = time.time()
    selection_sort(lst)
    end_time = time.time()
    ys.append((end_time - start_time) * 1000)

plt.scatter(xs, ys)

plt.title('Time of selection_sort(n)')
plt.xlabel('n')
plt.ylabel('time(ms)')

plt.show()
```
